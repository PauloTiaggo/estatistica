```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as stats
plt.style.use('ggplot')

fig, ax = plt.subplots()
mu = 0
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
ax.plot(x, stats.t.pdf(x, 10),
       'r-', lw=5, alpha=0.6, label='t de Student')
    
plt.legend()
plt.show()
```

### Comparação t de Studente $ \times$ Normal

```python
fig, ax = plt.subplots()
mu = 0
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
ax.plot(x, stats.t.pdf(x, 5),
       'r-', lw=1, alpha=0.6, label='t de Student')
#normal
ax.plot(x, stats.norm.pdf(x),
       'b-', lw=1, alpha=0.6, label='Normal')
    
plt.legend()
plt.show()
```
