Exercício 1: Desejamos coletar uma amostra de uma v.a. $X$ com desvio padrão igual a 15. Qual deve ser o tamanho da amostra para que, com 0,95 de probabilidade, a média amostral não difira da populacional por mais de 3 unidades?
<span/>

```python
# n = (z/epsilon)^2 * sigma^2
import numpy as np
from scipy import stats as stats

z = stats.norm.ppf(0.975)      #para 95% de confiança
print('z = ', round(z,2))
epsilon = 3   #o erro desejado
sigma = 15    #desvio-padrão da população suposto conhecido

n = (z/epsilon)**2 * sigma**2
print('Tamanho da amostra: ', n)
```

Exercício 2: Seja $X$ uma v.a. com distribuição normal de média desconhecida e variância igual a 36.<br/>
    
* Para uma amostra de tamanho 50, obtivemos média amostral $\overline{X} = 18,5$. Construa um intervalo de confiança com coeficientes de confiança 91%, 96% e 99% para a média populacional.
* Para uma confiança de 94%, construa intervalos de confiança supondo três tamanhos de amostra 25, 50 e 100 (admita que todos forneceram média amostral igual a 18,5).
* Comente sobre a precisão dos intervalos construídos em (a) e (b).
<span/>

```python
# n=50
# sigma = 6
# xbarra = 18.5
# a) 1-alfa = 91%
n=50
sigma=6
xbarra = 18.5
epsilon = stats.norm.ppf(0.955)*sigma/np.sqrt(n)
#Intervalo de confiança de 91%
# [Xbarra - epsilon; Xbarra + epsilon]
print('[', round(xbarra - epsilon,2), '; ', round(xbarra + epsilon,2), ']')
#
print('\n Calculando com o scipy stats:')
conf_int = stats.norm.interval(0.91, loc=xbarra, scale=sigma/np.sqrt(n))
print("\n intervalo de confiança 91%: ")
print(conf_int)
#
conf_int = stats.norm.interval(0.96, loc=xbarra, scale=sigma/np.sqrt(n))
print("\n intervalo de confiança 96%: ")
print(conf_int)
#
conf_int = stats.norm.interval(0.99, loc=xbarra, scale=sigma/np.sqrt(n))
print("\n intervalo de confiança 99%: ")
print(conf_int)
```

```python
#Para uma confiança de 94%, construa intervalos de confiança supondo três
# tamanhos de amostra 25, 50 e 100 (admita que todos forneceram média amostral igual a 18,5).
n=25
conf_int = stats.norm.interval(0.94, loc=xbarra, scale=sigma/np.sqrt(n))
print("\n intervalo de confiança 94% com n = : ", n)
print(conf_int)

n=50
conf_int = stats.norm.interval(0.94, loc=xbarra, scale=sigma/np.sqrt(n))
print("\n intervalo de confiança 94% com n = : ", n)
print(conf_int)

n=100
conf_int = stats.norm.interval(0.94, loc=xbarra, scale=sigma/np.sqrt(n))
print("\n intervalo de confiança 94% com n = : ", n)
print(conf_int)

n=10000
conf_int = stats.norm.interval(0.94, loc=xbarra, scale=sigma/np.sqrt(n))
print("\n intervalo de confiança 94% com n = : ", n)
print(conf_int)
```

Exercício 3. O intervalo [34,81 ; 36,38], com confiança 95% foi construído a partir de uma amostra de tamanho 25, para a média de uma população normal. O desvio padrão populacional é igual a 2.

* Qual o valor encontrado para a média dessa amostra?
* Se utilizássemos essa mesma amostra, mas com uma confiança de 90%, qual seria o novo intervalo de confiança?
<span/>

```python
#Intervalo de confiança: [Xbarra - erro; xbarra + erro]
# xbarra - erro = 34,81
# xbarra + erro = 36,38

# 2 * erro = 36,38 - 34,81
erro = (36.38 - 34.81)/2
epsilon = stats.norm.ppf(0.975)*2/np.sqrt(25)
print(erro, epsilon)
print('xbarra = ', 36.38 - erro)
```

```python
conf_int = stats.norm.interval(0.95, loc=35.6, scale=2/np.sqrt(25))
print("\n intervalo de confiança 90% com n = : ", 25)
print(conf_int)
```

Exercício 4. O tempo de permanência de contadores recém formados no primeiro emprego, em anos, foi estudado considerando um modelo normal com média e variância desconhecidas. Deseja-se estimar a média populacional. Para uma amostra de 15 profissionais, a média obtida foi de 2,7 anos e o desvio padrão foi de 1,4 anos. <br/>

* Encontre um intervalo para o tempo médio populacional de permanência com uma confiança de 90%.
* Refaça o item anterior, considerando que a amostra era formada por 150 profissionais.
<span/>

```python
n = 15
xbarra = 2.7
s = 1.4
gl = n-1 #graus de liberdade

erro = stats.t.ppf(0.95, df=gl)*s/np.sqrt(n)
print('erro: ', erro)
print('intervalo: ', xbarra-erro, xbarra+erro)

conf_int = stats.t.interval(0.90, df= gl, loc=xbarra, scale=s/np.sqrt(n))
print("\n intervalo de confiança 90% com n = : ", n)
print(conf_int)
```

```python
n = 150
xbarra = 2.7
s = 1.4
gl = n-1 #graus de liberdade

erro_t = stats.t.ppf(0.95, df=gl)*s/np.sqrt(n)
print('erro t de student: ', erro_t)
print('intervalo: ', round(xbarra-erro_t,2), round(xbarra+erro_t,2))

erro_n = stats.norm.ppf(0.95)*s/np.sqrt(n)
print('erro normal padrão: ', erro_n)
print('intervalo: ', round(xbarra-erro_n,2), round(xbarra+erro_n,2))
```

Exercício 5. Considere uma amostra com setenta observações proveniente de uma população com desvio padrão 5. Supondo que a média amostral seja 89, determine o intervalo com 95% de confiança para a média dessa população.
<span/>

```python
n = 70  # implica xbarra normal pelo TLC
sigma = 5 # desvio-padrão populacional
xbarra = 89
erro = 1.96 * sigma/np.sqrt(n)
print(round(erro,2))

print('intervalo de 95%:', xbarra-erro, xbarra+erro)
```
