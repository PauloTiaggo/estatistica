# Testes de hipóteses

* Introdução ao conceito de teste de hipóteses, testes de hipótese bilaterais e testes unilaterais 
* Teste de hipótese para a média e para a proporção de uma população, teste para a diferença entre duas populações
* Análise de variância

## Coméntários sobre os testes de hipóteses
* ### Interesse na decisão sobre a verdade ou não de uma hipótese específica. 
* ### Visa verificar se são verdadeiras as afirmações sobre os  parâmetros de uma ou mais populações. 
* ### Existem duas hipóteses: a hipótese nula ($H0$) e a hipótese alternativa ($H1$). 
* ### O objetivo do teste de hipóteses é <span style="font-color=red"> rejeitar </span> $H0$ com um nível de significância $\alpha$.

# Teste de hipóteses bilateral

Hipótese nula $H0$ : parâmetro $= \theta$ <br/>
Hipótese alternativa $H1$ : parâmetro $\neq \theta$ <br/>
Nível de significância: $(\alpha/2)\%$


Regiões de rejeição da hipótese nula dada pelo nível de significância do teste:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as stats
plt.style.use('ggplot')

fig, ax = plt.subplots()
ax.annotate('Rejeição de H0',
            xy=(2.2, 0.005), xycoords='data',
            xytext=(5.5, 0.15), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.03),
            horizontalalignment='right', verticalalignment='top', fontsize=20)

ax.annotate('Rejeição de H0',
            xy=(-2.2, 0.005), xycoords='data',
            xytext=(-3.5, 0.15), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.03),
            horizontalalignment='right', verticalalignment='top', fontsize=20)

plt.rcParams['figure.figsize'] = (15,9)
mean=0
std=1
x=np.linspace(-5, 5,1000)
iq=stats.norm(mean,std)
fig = plt.plot(x,iq.pdf(x),'b')

#calda esquerda
ltail=np.arange(-5,-1.96,0.1)
plt.fill_between(ltail,iq.pdf(ltail),color='r')

#calda direita
rtail=np.arange(1.96,5, 0.1)
plt.fill_between(rtail,iq.pdf(rtail),color='r')

#intervalo de confiança
xp = np.arange(-1.96,1.96, 0.1)
plt.fill_between(xp,iq.pdf(xp),color='w', label="95%")

locs, labels = plt.xticks()

plt.xticks([-6, -4, -1.96, 0, 1.96, 4,6], ['-6', '-4', '-1.96', 
                  '0', '$1.96 = z_{crítico}$', '4','6'], fontsize=16)

plt.legend()
plt.show()
```

# Testes de hipótese para a média de uma população
#### Nesse caso, o objetivo é testar a hipótese de que a média da população é igual ou não a um determinado valor especificado $\mu_0$.

#### Etapas para realização do teste de hipóteses:

* ##### Passo 1: Definição das hipóteses
<center> $H0: \mu = \mu_0$  </center>   </br>
<center> $H1: \mu \neq \mu_0$ </center>   </br>

* ##### Passo 2: Cálculo da estatística de teste - teste $z$ ou teste $t$?
* ##### Passo 3: Definição da região crítica - $1-\alpha$ de confiança
* ##### Passo 4: Avaliação do critério de rejeição: compara valor do teste com o valor crítico da distribuição
* ##### Passo 5: Conclusão: estabelece a conclusão conforme resultados do passo anterior.

## Caso 1: supondo que o $\sigma$ é conhecido

Estatística de teste <br/>

<span style="font-size:20pt">
<center> $z_{teste} = \frac{\overline{x}-\mu_0}{\frac{\sigma}{\sqrt{n}}}$ </center>
</span>

<span style="color:blue">
Exemplo 1. A vida média de uma amostra de 140 lâmpadas é de 1540 horas. Se $\mu$ e $\sigma = 130$ são, respectivamente, a vida média e o desvio padrão de todas as lâmpadas desse fabricante, teste a hipótese de que $\mu=1600$ horas contra a hipótese alternativa $\mu \neq 1600$ horas, com nível de significância $\alpha=0,05$.
    
</span> </br> 


<span style='color:red'>
Passo 1: definição das hipóteses </br>
</span>
<center> $H0: \mu = 1600$  </center>   </br>
<center> $H1: \mu \neq 1600$ </center>   </br>

<span style='color:red'>
Passo 2: cálculo da estatística
    </span>

```python
# Cálculo com a fórmula: z_teste = (xbarra - mu_0)/[sigma/raiz(n)]
xbarra = 1540
sigma = 130
mu_0 = 1600
n = 140

z_teste = (xbarra - mu_0)/(sigma/np.sqrt(140))
print(round(z_teste,2))
```

<span style='color:red'>
Passo 3: região crítica para teste bilateral com $95\%$
    </span>

```python
print('z_teste < -z_crítico = ', round(stats.norm.ppf(0.025),3))
print('ou \nz_teste > z_crítico = ', round(stats.norm.ppf(0.975),3))
```

<span style='color:red'> Passo 4: decisão   </br>

Rejeita-se $H0$, pois $-5.46 < -1.96$
</span>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as stats
plt.style.use('ggplot')

fig, ax = plt.subplots()
ax.annotate('Rejeição de H0',
            xy=(2.2, 0.005), xycoords='data',
            xytext=(5.5, 0.15), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.03),
            horizontalalignment='right', verticalalignment='top', fontsize=20)

ax.annotate('Rejeição de H0',
            xy=(-2.2, 0.005), xycoords='data',
            xytext=(-3.5, 0.15), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.03),
            horizontalalignment='right', verticalalignment='top', fontsize=20)

plt.rcParams['figure.figsize'] = (15,9)
mean=0
std=1
x=np.linspace(-5, 5,1000)
iq=stats.norm(mean,std)
fig = plt.plot(x,iq.pdf(x),'b')

#calda esquerda
ltail=np.arange(-5,-1.96,0.1)
plt.fill_between(ltail,iq.pdf(ltail),color='r')

#calda direita
rtail=np.arange(1.96,5, 0.1)
plt.fill_between(rtail,iq.pdf(rtail),color='r')

#intervalo de confiança
xp = np.arange(-1.96,1.96, 0.1)
plt.fill_between(xp,iq.pdf(xp),color='w', label="95%")

locs, labels = plt.xticks()

plt.xticks([-6, -4, -1.96, 0, 1.96, 4,6], ['-6', '$-5.461 = z_{teste}$''-4', '$-1.96 = -z_{crítico}$', 
                  '0', '1.96', '4','6'], fontsize=12)

plt.legend()
plt.show()
```

## Caso 2: supondo que o $\sigma$ é desconhecido

Estatística de teste <br/>

<span style="font-size:20pt">
<center> $t_{teste} = \frac{\overline{x}-\mu_0}{\frac{s}{\sqrt{n}}}$ </center>
</span>

<span style="color:blue">
Exemplo 2: O peso médio de uma amostra de 700 enlatados é de 500 gramas com desvio padrão de 20 gramas. O controle de qualidade da indústria quer testar se o peso médio de todas as latas fabricadas é de 475 gramas, com 99,5% de confiança.
    </span>

<span style='color:red'>
Passo 1: definição das hipóteses </br>
</span>
<center> $H0: \mu = 475$  </center>   </br>
<center> $H1: \mu \neq 475$ </center>   </br>

<span style='color:red'>
Passo 2: cálculo da estatística
    </span>

```python
# Cálculo com a fórmula: t_teste = (xbarra - mu_0)/[s/raiz(n)]
xbarra = 500
s = 20
mu_0 = 475
n = 700
SE = s/np.sqrt(700)
t_teste = (xbarra - mu_0)/SE
print(SE)
print(round(t_teste,3))
```

# Teste de hipótese para proporção
### Uma população
Usado para avaliar afirmações sobre a proporção de uma população

Exemplo: <br/>
<img src="exemplo_teste_hipotese_proporcao.jpg">

```python
from statsmodels.stats.proportion import proportions_ztest
n = 400 #número de observações
X = 320 #número de sucessos (80%)
alpha = 0.1 #significância

#H0: p  = 0.95 (Valor a ser testado)
#H1: p != 0.95

#####################
## Usando a formula
## z_teste = (phat - p0)/ raiz(phat*(1-phat)/n)
phat = 320/400
p0 = 0.95

z_teste = (phat - p0)/np.sqrt(phat*(1-phat)/n)
print("Z_teste pela fórmula:" , round(z_teste,2))

stat, pval = proportions_ztest(X, n, 0.95)
print('Z_teste scipy: {0:0.3f}'.format(stat))
```

### z-crítico para 90% de confiança ($\alpha = 10\%$)

```python
zcritico = stats.norm.ppf(0.95)
print(zcritico)
```

### Resultado: rejeita a hipótese nula, pois z_teste < -z_crítico

# Exemplos

```python
# generate gaussian data samples
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
# seed the random number generator
seed(1)
# generate two sets of univariate observations
data1 = 5 * randn(100) + 50
data2 = 5 * randn(100) + 51
# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))

# compare samples
stat, p = stats.ttest_ind(data1, data2)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

# Teste de hipóteses para proporção populacional

### Exemplo 1: uma população </br>
Em uma amostra de tamanho $n=500$ foram observados $X=410$ sucessos. Deseja-se testar se o valor da proporção populacional $p$ é igual ou maior que $p_0 = 0.80$, com 95% de confiança.

```python
from statsmodels.stats.proportion import proportions_ztest

# sifnificância
significance = 0.05

# Amostra
sample_success = 410
sample_size = 500

# H0 =  80%
null_hypothesis = 0.80

# Teste unilateral à esquerda >> alternative='smaller'
# Teste unilateral à direita >> alternative= 'larger'
# teste bilateral alternative='two-sided'
stat, p_value = proportions_ztest(count=sample_success, nobs=sample_size, value=null_hypothesis, alternative='larger')

# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

if p_value > significance:
   print ("Fail to reject the null hypothesis - we have nothing else to say")
else:
   print ("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

### Exemplo 2: 2 amostras para proporção
Nesse exemplo temos duas amostras: uma com $n_1 = 500$ e $X_1=410$ sucessos e outra com $n_2=400$ e sucesso $X_2=379$. As proporções populacionais podem ser consideradas iguais? </br>

$$
H0: p_1 = p_2 \\
H1: p_1 \neq p_2
$$

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

# can we assume anything from our sample
significance = 0.025

# our samples - 82% are good in one, and ~79% are good in the other
# note - the samples do not need to be the same size
sample_success_a, sample_size_a = (410, 500)
sample_success_b, sample_size_b = (379, 400)

# check our sample against Ho for Ha != Ho
successes = np.array([sample_success_a, sample_success_b])
samples = np.array([sample_size_a, sample_size_b])

# note, no need for a Ho value here - it's derived from the other parameters
stat, p_value = proportions_ztest(count=successes, nobs=samples,  alternative='two-sided')

# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

if p_value > significance:
   print ("Fail to reject the null hypothesis - we have nothing else to say")
else:
   print ("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

# Teste de hipóteses para duas variâncias

```python
import numpy as np

#Data
x = [18, 19, 22, 25, 27, 28, 41, 45, 51, 55]
y = [14, 15, 15, 17, 18, 22, 25, 25, 27, 34]

#define F-test function
def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    f_teste = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
    dfn = x.size-1 #define degrees of freedom numerator 
    dfd = y.size-1 #define degrees of freedom denominator 
    p = 1-stats.f.cdf(f_teste, dfn, dfd) #find p-value of F test statistic 
    return f_teste, p

#perform F-test
f_test(x, y)
```

# Métodos de Análise das Relações de Múltiplas Variáveis

```python
# Analysis of Variance test
from numpy.random import seed
from numpy.random import randn
from scipy.stats import f_oneway
# seed the random number generator
seed(1)
# generate three independent samples
data1 = 5 * randn(100) + 54
data2 = 5 * randn(100) + 50
data3 = 5 * randn(100) + 52
stat, p = f_oneway(data1, data2, data3)

alunos = [25, 10, 14, 14, 20, 27, 19, 22, 19, 14, 15, 17, 18]
executivos = [15, 18, 13, 15, 14, 9, 9, 10, 11, 13, 22, 8, 10]
funcionarios = [25, 29, 28, 27, 25, 28, 29, 27, 26, 28, 21, 14, 24]
stat, p = f_oneway(alunos, executivos, funcionarios)
# compare samples

print('Statistics=%.3f, p=%.10f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```
