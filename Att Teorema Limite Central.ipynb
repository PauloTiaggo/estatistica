```python
import math
import statistics
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import kstest
```

```python
def TCL_randn(n):
  np.random.seed(1)
  x = []
  for i in range(50):
    i = np.mean(np.random.randn(n))
    x.append(i)

  m = np.mean(x)
  o = np.std(x)
  plt.hist(x, cumulative=False, density=True, bins=10, color='crimson')

  y = np.linspace(m - 3*o, m + 3*o, 100)
  plt.plot(y, stats.norm.pdf(y, m, o), color='black')
  plt.title(f'N={n}')
  plt.show()
  print(stats.shapiro(x))
```

O teste de hip√≥tese Shapiro-Wilk retorna a estat√≠stica de teste W e o p-value. Se o resultado p-value √© maior que 0,05 conclui-se que a distribui√ß√£o normal √© adequada. Significa dizer que as 6 amostras aleat√≥rias geradas acima s√£o normalmente distribuidas, portanto n√£o h√° indicios suficientes para rejeitar a hip√≥tese nula.

```python
def TCL_exponential(n):
  np.random.seed(8)
  x = []
  for i in range(50):
    i = np.mean(np.random.exponential(1, n))
    x.append(i)

  m = np.mean(x)
  o = np.std(x)
  plt.hist(x, cumulative=False, density=True, bins=10, color='mediumblue')

  y = np.linspace(m - 3*o, m + 3*o, 100)
  plt.plot(y, stats.norm.pdf(y, m, o), color='black')
  plt.title(f'N={n}')
  plt.show()
  print(stats.shapiro(x))
```

```python
TCL_exponential(5)
```
